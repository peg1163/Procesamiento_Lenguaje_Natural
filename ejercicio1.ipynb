```python
ejemplos = [
    "el gato corre",
    "el perro ladra",
    "la casa es grande",
    "el gato duerme mucho",
    "el perro come comida",
    "la ni√±a juega con el perro"
]
tokens = []
for i in ejemplos:
    tokens.extend(i.lower().split())
    
vocab = sorted(set(tokens))

token_unk = "<unk>"
if token_unk not in vocab:
    vocab.append(token_unk)

dic = {w: i for i, w in enumerate(vocab)}

def translate(frase):
    tokens = frase.lower().split()
    unk_idx = dic[token_unk]
    return [dic.get(t, unk_idx) for t in tokens]
```

```python
frase = "el perro volador corre loco"
print(frase)
print(translate(frase))
```
